```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in all the validation_results.log files from the different experiments
adam = pd.read_csv('./adam/validation_results.log')
adamw = pd.read_csv('./adamw/validation_results.log')
sgd = pd.read_csv('./sgd/validation_results.log')
```

```python
def ema_smooth(x, alpha=0.2):
    smoothed = []
    last = x[0]
    for v in x:
        last = alpha * v + (1 - alpha) * last
        smoothed.append(last)
    return smoothed
```

```python
# Plotting all the validation loss curves for comparison
# Use vaLidation step as x-axis and validation loss as y-axis
plt.figure(figsize=(12, 8))
plt.plot(adam['step'], adam[' val_loss'], label='Adam', alpha=0.45, color='teal')
plt.plot(adam['step'], ema_smooth(adam[' val_loss']), label='Adam (smoothed)', color='teal')
plt.plot(adamw['step'], adamw[' val_loss'], label='AdamW', alpha=0.45, color='orange')
plt.plot(adamw['step'], ema_smooth(adamw[' val_loss']), label='AdamW (smoothed)', color='orange')
plt.plot(sgd['step'], sgd[' val_loss'], label='SGD', alpha=0.45, color='purple')
plt.plot(sgd['step'], ema_smooth(sgd[' val_loss']), label='SGD (smoothed)', color='purple')
plt.xlabel('Validation Step')
plt.ylabel('Validation Loss')
plt.title('Validation Loss Comparison Across Different Optimizer Setups')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plotting all the validation accuracy curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(adam['step'], adam[' val_acc'], label='Adam', alpha=0.45, color='teal')
plt.plot(adam['step'], ema_smooth(adam[' val_acc']), label='Adam (smoothed)', color='teal')
plt.plot(adamw['step'], adamw[' val_acc'], label='AdamW', alpha=0.45, color='orange')
plt.plot(adamw['step'], ema_smooth(adamw[' val_acc']), label='AdamW (smoothed)', color='orange')
plt.plot(sgd['step'], sgd[' val_acc'], label='SGD', alpha=0.45, color='purple')
plt.plot(sgd['step'], ema_smooth(sgd[' val_acc']), label='SGD (smoothed)', color='purple')
plt.xlabel('Validation Step')
plt.ylabel('Validation Accuracy')
plt.title('Validation Accuracy Comparison Across Different Optimizer Setups')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plotting all the validation accuracy at the last character curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(adam['step'], adam[' last_char_val_acc'], label='Adam', alpha=0.3, color='teal')
plt.plot(adam['step'], ema_smooth(adam[' last_char_val_acc']), label='Adam (smoothed)', color='teal')
plt.plot(adamw['step'], adamw[' last_char_val_acc'], label='AdamW', alpha=0.3, color='orange')
plt.plot(adamw['step'], ema_smooth(adamw[' last_char_val_acc']), label='AdamW (smoothed)', color='orange')
plt.plot(sgd['step'], sgd[' last_char_val_acc'], label='SGD', alpha=0.3, color='purple')
plt.plot(sgd['step'], ema_smooth(sgd[' last_char_val_acc']), label='SGD (smoothed)', color='purple')
plt.xlabel('Validation Step')
plt.ylabel('Validation Accuracy at Last Character')
plt.title('Validation Accuracy at Last Character Comparison Across Different Optimizer Setups')
plt.legend()
plt.grid()
plt.show()
```
