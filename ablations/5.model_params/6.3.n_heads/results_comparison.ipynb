```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in all the validation_results.log files from the different experiments
n_heads_2 = pd.read_csv('./n_heads_2/validation_results.log')
n_heads_4 = pd.read_csv('./n_heads_4/validation_results.log')
n_heads_8= pd.read_csv('./n_heads_8/validation_results.log')
```

```python
def ema_smooth(x, alpha=0.2):
    smoothed = []
    last = x[0]
    for v in x:
        last = alpha * v + (1 - alpha) * last
        smoothed.append(last)
    return smoothed
```

```python
# Plotting all the validation loss curves for comparison
# Use vaLidation step as x-axis and validation loss as y-axis
plt.figure(figsize=(12, 8))

plt.plot(n_heads_2['step'], n_heads_2[' val_loss'], label='n_heads_2', alpha=0.15, color='teal')
plt.plot(n_heads_2['step'], ema_smooth(n_heads_2[' val_loss']), label='n_heads_2 (smoothed)', color='teal')

plt.plot(n_heads_4['step'], n_heads_4[' val_loss'], label='n_heads_4', alpha=0.15, color='orange')
plt.plot(n_heads_4['step'], ema_smooth(n_heads_4[' val_loss']), label='n_heads_4 (smoothed)', color='orange')

plt.plot(n_heads_8['step'], n_heads_8[' val_loss'], label='n_heads_8', alpha=0.15, color='purple')
plt.plot(n_heads_8['step'], ema_smooth(n_heads_8[' val_loss']), label='n_heads_8 (smoothed)', color='purple')

plt.xlabel('Validation Step')
plt.ylabel('Validation Loss')
plt.yscale('log')
plt.title('Validation Loss Comparison Across Different Number of Attention Heads')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plotting all the validation accuracy curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(n_heads_2['step'], n_heads_2[' val_acc'], label='n_heads_2', alpha=0.15, color='teal')
plt.plot(n_heads_2['step'], ema_smooth(n_heads_2[' val_acc']), label='n_heads_2 (smoothed)', color='teal')

plt.plot(n_heads_4['step'], n_heads_4[' val_acc'], label='n_heads_4', alpha=0.15, color='orange')
plt.plot(n_heads_4['step'], ema_smooth(n_heads_4[' val_acc']), label='n_heads_4 (smoothed)', color='orange')

plt.plot(n_heads_8['step'], n_heads_8[' val_acc'], label='n_heads_8', alpha=0.15, color='purple')
plt.plot(n_heads_8['step'], ema_smooth(n_heads_8[' val_acc']), label='n_heads_8 (smoothed)', color='purple')

plt.xlabel('Validation Step')
plt.ylabel('Validation Accuracy')
plt.yscale('log')
plt.title('Validation Accuracy Comparison Across Different Number of Attention Heads')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plotting all the validation accuracy at the last character curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(n_heads_2['step'], n_heads_2[' last_char_val_acc'], label='n_heads_2', alpha=0.15, color='teal')
plt.plot(n_heads_2['step'], ema_smooth(n_heads_2[' last_char_val_acc']), label='n_heads_2 (smoothed)', color='teal')

plt.plot(n_heads_4['step'], n_heads_4[' last_char_val_acc'], label='n_heads_4', alpha=0.15, color='orange')
plt.plot(n_heads_4['step'], ema_smooth(n_heads_4[' last_char_val_acc']), label='n_heads_4 (smoothed)', color='orange')

plt.plot(n_heads_8['step'], n_heads_8[' last_char_val_acc'], label='n_heads_8', alpha=0.15, color='purple')
plt.plot(n_heads_8['step'], ema_smooth(n_heads_8[' last_char_val_acc']), label='n_heads_8 (smoothed)', color='purple')

plt.xlabel('Validation Step')
plt.ylabel('Validation Accuracy at Last Character')
plt.title('Validation Accuracy at Last Character Comparison Across Different Number of Attention Heads')
plt.legend()
plt.grid()
plt.show()
```
