```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in all the validation_results.log files from the different experiments
d_model_256 = pd.read_csv('./d_model_256/validation_results.log')
d_model_512 = pd.read_csv('./d_model_512/validation_results.log')
d_model_128= pd.read_csv('./d_model_128/validation_results.log')
```

```python
def ema_smooth(x, alpha=0.2):
    smoothed = []
    last = x[0]
    for v in x:
        last = alpha * v + (1 - alpha) * last
        smoothed.append(last)
    return smoothed
```

```python
# Plotting all the validation loss curves for comparison
# Use vaLidation step as x-axis and validation loss as y-axis
plt.figure(figsize=(12, 8))

plt.plot(d_model_128['step'], d_model_128[' val_loss'], label='d_model_128', alpha=0.15, color='teal')
plt.plot(d_model_128['step'], ema_smooth(d_model_128[' val_loss']), label='d_model_128 (smoothed)', color='teal')

plt.plot(d_model_256['step'], d_model_256[' val_loss'], label='d_model_256', alpha=0.15, color='orange')
plt.plot(d_model_256['step'], ema_smooth(d_model_256[' val_loss']), label='d_model_256 (smoothed)', color='orange')

plt.plot(d_model_512['step'], d_model_512[' val_loss'], label='d_model_512', alpha=0.15, color='purple')
plt.plot(d_model_512['step'], ema_smooth(d_model_512[' val_loss']), label='d_model_512 (smoothed)', color='purple')

plt.xlabel('Validation Step')
plt.ylabel('Validation Loss')
plt.title('Validation Loss Comparison Across Different d_model Setups')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plotting all the validation accuracy curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(d_model_128['step'], d_model_128[' val_acc'], label='d_model_128', alpha=0.15, color='teal')
plt.plot(d_model_128['step'], ema_smooth(d_model_128[' val_acc']), label='d_model_128 (smoothed)', color='teal')

plt.plot(d_model_256['step'], d_model_256[' val_acc'], label='d_model_256', alpha=0.15, color='orange')
plt.plot(d_model_256['step'], ema_smooth(d_model_256[' val_acc']), label='d_model_256 (smoothed)', color='orange')

plt.plot(d_model_512['step'], d_model_512[' val_acc'], label='d_model_512', alpha=0.15, color='purple')
plt.plot(d_model_512['step'], ema_smooth(d_model_512[' val_acc']), label='d_model_512 (smoothed)', color='purple')

plt.xlabel('Validation Step')
plt.ylabel('Validation Accuracy')
plt.title('Validation Accuracy Comparison Across Different d_model Setups')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plotting all the validation accuracy at the last character curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(d_model_128['step'], d_model_128[' last_char_val_acc'], label='d_model_128', alpha=0.15, color='teal')
plt.plot(d_model_128['step'], ema_smooth(d_model_128[' last_char_val_acc']), label='d_model_128 (smoothed)', color='teal')

plt.plot(d_model_256['step'], d_model_256[' last_char_val_acc'], label='d_model_256', alpha=0.15, color='orange')
plt.plot(d_model_256['step'], ema_smooth(d_model_256[' last_char_val_acc']), label='d_model_256 (smoothed)', color='orange')

plt.plot(d_model_512['step'], d_model_512[' last_char_val_acc'], label='d_model_512', alpha=0.15, color='purple')
plt.plot(d_model_512['step'], ema_smooth(d_model_512[' last_char_val_acc']), label='d_model_512 (smoothed)', color='purple')

plt.xlabel('Validation Step')
plt.ylabel('Validation Accuracy at Last Character')
plt.title('Validation Accuracy at Last Character Comparison Across Different d_model Setups')
plt.legend()
plt.grid()
plt.show()
```
