```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in all the validation_results.log files from the different experiments
n_layer_4 = pd.read_csv('./n_layers_4/validation_results.log')
n_layer_6 = pd.read_csv('./n_layers_6/validation_results.log')
n_layer_8= pd.read_csv('./n_layers_8/validation_results.log')
```

```python
def ema_smooth(x, alpha=0.2):
    smoothed = []
    last = x[0]
    for v in x:
        last = alpha * v + (1 - alpha) * last
        smoothed.append(last)
    return smoothed
```

```python
# Plotting all the validation loss curves for comparison
# Use vaLidation step as x-axis and validation loss as y-axis
plt.figure(figsize=(12, 8))

plt.plot(n_layer_4['step'], n_layer_4[' val_loss'], alpha=0.15, color='red')
plt.plot(n_layer_4['step'], ema_smooth(n_layer_4[' val_loss']), label='4 Layers', color='red')
plt.plot(n_layer_6['step'], n_layer_6[' val_loss'], alpha=0.15, color='orange')
plt.plot(n_layer_6['step'], ema_smooth(n_layer_6[' val_loss']), label='6 Layers', color='orange')
plt.plot(n_layer_8['step'], n_layer_8[' val_loss'], alpha=0.15, color='green')
plt.plot(n_layer_8['step'], ema_smooth(n_layer_8[' val_loss']), label='8 Layers', color='green')
plt.xlabel('Validation Step')
plt.ylabel('Validation Loss')
plt.yscale('log')
plt.title('No. of Layers (n_layers)')
plt.legend()
plt.grid()
plt.savefig('n_layers.jpg')
plt.show()
```

```python
# Plotting all the validation accuracy curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(n_layer_4['step'], n_layer_4[' val_acc'], alpha=0.15, color='red')
plt.plot(n_layer_4['step'], ema_smooth(n_layer_4[' val_acc']), label='4 Layers', color='red')
plt.plot(n_layer_6['step'], n_layer_6[' val_acc'], alpha=0.15, color='orange')
plt.plot(n_layer_6['step'], ema_smooth(n_layer_6[' val_acc']), label='6 Layers', color='orange')
plt.plot(n_layer_8['step'], n_layer_8[' val_acc'], alpha=0.15, color='green')
plt.plot(n_layer_8['step'], ema_smooth(n_layer_8[' val_acc']), label='8 Layers', color='green')
plt.xlabel('Validation Step')
plt.ylabel('Validation Accuracy')
plt.title('No. of Layers (n_layers)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plotting all the validation accuracy at the last character curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(n_layer_4['step'], n_layer_4[' last_char_val_acc'], alpha=0.15, color='red')
plt.plot(n_layer_4['step'], ema_smooth(n_layer_4[' last_char_val_acc']), label='4 Layers', color='red')
plt.plot(n_layer_6['step'], n_layer_6[' last_char_val_acc'], alpha=0.15, color='orange')
plt.plot(n_layer_6['step'], ema_smooth(n_layer_6[' last_char_val_acc']), label='6 Layers', color='orange')
plt.plot(n_layer_8['step'], n_layer_8[' last_char_val_acc'], alpha=0.15, color='green')
plt.plot(n_layer_8['step'], ema_smooth(n_layer_8[' last_char_val_acc']), label='8 Layers', color='green')
plt.xlabel('Validation Step')
plt.ylabel('Validation Accuracy at Last Character')
plt.title('No. of Layers (n_layers)')
plt.legend()
plt.grid()
plt.show()
```
