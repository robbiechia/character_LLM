```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in all the validation_results.log files from the different experiments
n_layer_4 = pd.read_csv('./n_layers_4/validation_results.log')
n_layer_6 = pd.read_csv('./n_layers_6/validation_results.log')
n_layer_8= pd.read_csv('./n_layers_8/validation_results.log')
```

```python
def ema_smooth(x, alpha=0.2):
    smoothed = []
    last = x[0]
    for v in x:
        last = alpha * v + (1 - alpha) * last
        smoothed.append(last)
    return smoothed
```

```python
# Plotting all the validation loss curves for comparison
# Use vaLidation step as x-axis and validation loss as y-axis
plt.figure(figsize=(12, 8))

plt.plot(n_layer_4[' val_time'], n_layer_4[' val_loss'], label='4 Layers', alpha=0.15, color='teal')
plt.plot(n_layer_4[' val_time'], ema_smooth(n_layer_4[' val_loss']), label='4 Layers (smoothed)', color='teal')
plt.plot(n_layer_6[' val_time'], n_layer_6[' val_loss'], label='6 Layers', alpha=0.15, color='orange')
plt.plot(n_layer_6[' val_time'], ema_smooth(n_layer_6[' val_loss']), label='6 Layers (smoothed)', color='orange')
plt.plot(n_layer_8[' val_time'], n_layer_8[' val_loss'], label='8 Layers', alpha=0.15, color='purple')
plt.plot(n_layer_8[' val_time'], ema_smooth(n_layer_8[' val_loss']), label='8 Layers (smoothed)', color='purple')

plt.xlabel('Validation Time (s)')
plt.ylabel('Validation Loss')
plt.yscale('log')
plt.title('Validation Loss Comparison Across Different Attention Mechanism Setups')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plotting all the validation accuracy curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(n_layer_4[' val_time'], n_layer_4[' val_acc'], label='n_layer_4', alpha=0.15, color='teal')
plt.plot(n_layer_4[' val_time'], ema_smooth(n_layer_4[' val_acc']), label='n_layer_4 (smoothed)', color='teal')
plt.plot(n_layer_6[' val_time'], n_layer_6[' val_acc'], label='n_layer_6', alpha=0.15, color='orange')
plt.plot(n_layer_6[' val_time'], ema_smooth(n_layer_6[' val_acc']), label='n_layer_6 (smoothed)', color='orange')
plt.plot(n_layer_8[' val_time'], n_layer_8[' val_acc'], label='n_layer_8', alpha=0.15, color='purple')
plt.plot(n_layer_8[' val_time'], ema_smooth(n_layer_8[' val_acc']), label='n_layer_8 (smoothed)', color='purple')
plt.xlabel('Validation Time (s)')
plt.ylabel('Validation Accuracy')
plt.title('Validation Accuracy Comparison Across Different Attention Mechanisms')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plotting all the validation accuracy at the last character curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(n_layer_4[' val_time'], n_layer_4[' last_char_val_acc'], label='n_layer_4', alpha=0.15, color='teal')
plt.plot(n_layer_4[' val_time'], ema_smooth(n_layer_4[' last_char_val_acc']), label='n_layer_4 (smoothed)', color='teal')
plt.plot(n_layer_6[' val_time'], n_layer_6[' last_char_val_acc'], label='n_layer_6', alpha=0.15, color='orange')
plt.plot(n_layer_6[' val_time'], ema_smooth(n_layer_6[' last_char_val_acc']), label='n_layer_6 (smoothed)', color='orange')
plt.plot(n_layer_8[' val_time'], n_layer_8[' last_char_val_acc'], label='n_layer_8', alpha=0.15, color='purple')
plt.plot(n_layer_8[' val_time'], ema_smooth(n_layer_8[' last_char_val_acc']), label='n_layer_8 (smoothed)', color='purple')
plt.xlabel('Validation Time (s)')
plt.ylabel('Validation Accuracy at Last Character')
plt.title('Validation Accuracy at Last Character Comparison Across Different Attention Mechanism Setups')
plt.legend()
plt.grid()
plt.show()
```
