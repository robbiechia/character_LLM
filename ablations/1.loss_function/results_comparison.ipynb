```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in all the validation_results.log files from the different experiments
every_position_with_aux = pd.read_csv('./every_position_with_aux/validation_results.log')
every_position_without_aux = pd.read_csv('./every_position_without_aux/validation_results.log')
last_position_with_aux = pd.read_csv('./last_position_with_aux/validation_results.log')
last_position_without_aux = pd.read_csv('./last_position_without_aux/validation_results.log')
```

```python
# Plotting all the validation loss curves for comparison
# Use vaLidation step as x-axis and validation loss as y-axis
plt.figure(figsize=(12, 8))
plt.plot(every_position_with_aux['step'], every_position_with_aux['validation_loss'], label='Every Position with Aux Loss')
plt.plot(every_position_without_aux['step'], every_position_without_aux['validation_loss'], label='Every Position without Aux Loss')
plt.plot(last_position_with_aux['step'], last_position_with_aux['validation_loss'], label='Last Position with Aux Loss')
plt.plot(last_position_without_aux['step'], last_position_without_aux['validation_loss'], label='Last Position without Aux Loss')
plt.xlabel('Validation Step')
plt.ylabel('Validation Loss')
plt.title('Validation Loss Comparison Across Different Loss Function Setups')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plotting all the validation accuracy curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(every_position_with_aux['step'], every_position_with_aux['val_acc'], label='Every Position with Aux Loss')
plt.plot(every_position_without_aux['step'], every_position_without_aux['val_acc'], label='Every Position without Aux Loss')
plt.plot(last_position_with_aux['step'], last_position_with_aux['val_acc'], label='Last Position with Aux Loss')
plt.plot(last_position_without_aux['step'], last_position_without_aux['val_acc'], label='Last Position without Aux Loss')
plt.xlabel('Validation Step')
plt.ylabel('Validation Accuracy')
plt.title('Validation Accuracy Comparison Across Different Loss Function Setups')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plotting all the validation accuracy at the last character curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(every_position_with_aux['step'], every_position_with_aux['last_char_val_acc'], label='Every Position with Aux Loss')
plt.plot(every_position_without_aux['step'], every_position_without_aux['last_char_val_acc'], label='Every Position without Aux Loss')
plt.plot(last_position_with_aux['step'], last_position_with_aux['last_char_val_acc'], label='Last Position with Aux Loss')
plt.plot(last_position_without_aux['step'], last_position_without_aux['last_char_val_acc'], label='Last Position without Aux Loss')
plt.xlabel('Validation Step')
plt.ylabel('Validation Accuracy at Last Character')
plt.title('Validation Accuracy at Last Character Comparison Across Different Loss Function Setups')
plt.legend()
plt.grid()
plt.show()
```
