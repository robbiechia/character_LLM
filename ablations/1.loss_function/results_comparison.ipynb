```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in all the validation_results.log files from the different experiments
every_position_with_aux = pd.read_csv('./every_position_with_aux/validation_results.log')
every_position_without_aux = pd.read_csv('./every_position_without_aux/validation_results.log')
last_position_with_aux = pd.read_csv('./last_position_with_aux/validation_results.log')
last_position_without_aux = pd.read_csv('./last_position_without_aux/validation_results.log')
```

```python
def ema_smooth(x, alpha=0.2):
    smoothed = []
    last = x[0]
    for v in x:
        last = alpha * v + (1 - alpha) * last
        smoothed.append(last)
    return smoothed
```

```python
# Plotting all the validation loss curves for comparison
# Use vaLidation step as x-axis and validation loss as y-axis
plt.figure(figsize=(12, 8))
plt.plot(every_position_with_aux['step'], ema_smooth(every_position_with_aux[' val_loss']), label='Every Position with Aux Loss', color='red')
plt.plot(every_position_with_aux['step'], every_position_with_aux[' val_loss'], alpha=0.3, color='red')
plt.plot(every_position_without_aux['step'], ema_smooth(every_position_without_aux[' val_loss']), label='Every Position without Aux Loss', color='orange')
plt.plot(every_position_without_aux['step'], every_position_without_aux[' val_loss'], alpha=0.3, color='orange')
plt.plot(last_position_with_aux['step'], ema_smooth(last_position_with_aux[' val_loss']), label='Last Position with Aux Loss', color='green')
plt.plot(last_position_with_aux['step'], last_position_with_aux[' val_loss'], alpha=0.3, color='green')
plt.plot(last_position_without_aux['step'], ema_smooth(last_position_without_aux[' val_loss']), label='Last Position without Aux Loss', color='blue')
plt.plot(last_position_without_aux['step'], last_position_without_aux[' val_loss'], alpha=0.3, color='blue')
plt.xlabel('Validation Step')
plt.ylabel('Validation Loss')
plt.title('Loss Function')
plt.legend()
plt.grid()
plt.savefig('loss_function.jpg')
plt.show()
```

```python
# Plotting all the validation accuracy curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(every_position_with_aux['step'], ema_smooth(every_position_with_aux[' val_acc']), label='Every Position with Aux Loss', color='red')
plt.plot(every_position_with_aux['step'], every_position_with_aux[' val_acc'], alpha=0.3, color='red')
plt.plot(every_position_without_aux['step'], ema_smooth(every_position_without_aux[' val_acc']), label='Every Position without Aux Loss', color='orange')
plt.plot(every_position_without_aux['step'], every_position_without_aux[' val_acc'], alpha=0.3, color='orange')
plt.plot(last_position_with_aux['step'], ema_smooth(last_position_with_aux[' val_acc']), label='Last Position with Aux Loss', color='green')
plt.plot(last_position_with_aux['step'], last_position_with_aux[' val_acc'], alpha=0.3, color='green')
plt.plot(last_position_without_aux['step'], ema_smooth(last_position_without_aux[' val_acc']), label='Last Position without Aux Loss', color='blue')
plt.plot(last_position_without_aux['step'], last_position_without_aux[' val_acc'], alpha=0.3, color='blue')
plt.xlabel('Validation Step')
plt.ylabel('Validation Accuracy')
plt.title('Loss Function')
plt.legend()
plt.grid()
plt.show()
```
