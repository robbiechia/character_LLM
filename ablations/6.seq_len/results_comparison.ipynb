```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in all the validation_results.log files from the different experiments
seq_len_64 = pd.read_csv('./seq_len_64/validation_results.log')
seq_len_128 = pd.read_csv('./seq_len_128/validation_results.log')
seq_len_256 = pd.read_csv('./seq_len_256/validation_results.log')
```

```python
def ema_smooth(x, alpha=0.2):
    smoothed = []
    last = x[0]
    for v in x:
        last = alpha * v + (1 - alpha) * last
        smoothed.append(last)
    return smoothed
```

```python
# Plotting all the validation loss curves for comparison
# Use vaLidation step as x-axis and validation loss as y-axis
plt.figure(figsize=(12, 8))

plt.plot(seq_len_64['step'], seq_len_64[' val_loss'], alpha=0.15, color='red')
plt.plot(seq_len_64['step'], ema_smooth(seq_len_64[' val_loss']), label='seq_len: 64', color='red')

plt.plot(seq_len_128['step'], seq_len_128[' val_loss'], alpha=0.15, color='orange')
plt.plot(seq_len_128['step'], ema_smooth(seq_len_128[' val_loss']), label='seq_len: 128', color='orange')

plt.plot(seq_len_256['step'], seq_len_256[' val_loss'], alpha=0.15, color='green')
plt.plot(seq_len_256['step'], ema_smooth(seq_len_256[' val_loss']), label='seq_len: 256', color='green')
plt.yscale('log')
plt.xlabel('Validation Step')
plt.ylabel('Validation Loss')
plt.title('Sequence Length L')
plt.legend()
plt.grid()
plt.savefig('seq_len.jpg')
plt.show()
```

```python
# Plotting all the validation accuracy curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(seq_len_64['step'], seq_len_64[' val_acc'], alpha=0.15, color='red')
plt.plot(seq_len_64['step'], ema_smooth(seq_len_64[' val_acc']), label='seq_len: 64', color='red')

plt.plot(seq_len_128['step'], seq_len_128[' val_acc'], alpha=0.15, color='orange')
plt.plot(seq_len_128['step'], ema_smooth(seq_len_128[' val_acc']), label='seq_len: 128', color='orange')

plt.plot(seq_len_256['step'], seq_len_256[' val_acc'], alpha=0.15, color='green')
plt.plot(seq_len_256['step'], ema_smooth(seq_len_256[' val_acc']), label='seq_len: 256', color='green')
plt.xlabel('Validation Step')
plt.ylabel('Validation Accuracy')
plt.title('Sequence Length L')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plotting all the validation accuracy at the last character curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(seq_len_64['step'], seq_len_64[' last_char_val_acc'], alpha=0.15, color='red')
plt.plot(seq_len_64['step'], ema_smooth(seq_len_64[' last_char_val_acc']), label='seq_len: 64', color='red')

plt.plot(seq_len_128['step'], seq_len_128[' last_char_val_acc'], alpha=0.15, color='orange')
plt.plot(seq_len_128['step'], ema_smooth(seq_len_128[' last_char_val_acc']), label='seq_len: 128', color='orange')

plt.plot(seq_len_256['step'], seq_len_256[' last_char_val_acc'], alpha=0.15, color='green')
plt.plot(seq_len_256['step'], ema_smooth(seq_len_256[' last_char_val_acc']), label='seq_len: 256', color='green')

plt.xlabel('Validation Step')
plt.ylabel('Validation Accuracy at Last Character')
plt.title('Sequence Length L')
plt.legend()
plt.grid()
plt.show()
```
