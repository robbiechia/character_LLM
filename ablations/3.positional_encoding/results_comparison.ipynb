```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in all the validation_results.log files from the different experiments
alibi = pd.read_csv('./alibi/validation_results.log')
learned = pd.read_csv('./learned/validation_results.log')
none = pd.read_csv('./none/validation_results.log')
rotary = pd.read_csv('./rotary/validation_results.log')
sinusoidal = pd.read_csv('./sinusoidal/validation_results.log')
```

```python
def ema_smooth(x, alpha=0.2):
    smoothed = []
    last = x[0]
    for v in x:
        last = alpha * v + (1 - alpha) * last
        smoothed.append(last)
    return smoothed
```

```python
# Plotting all the validation loss curves for comparison
# Use vaLidation step as x-axis and validation loss as y-axis
plt.figure(figsize=(12, 8))

plt.plot(alibi['step'], alibi[' val_loss'], alpha=0.15, color='red')
plt.plot(alibi['step'], ema_smooth(alibi[' val_loss']), label='Alibi', color='red')

plt.plot(learned['step'], learned[' val_loss'], alpha=0.15, color='orange')
plt.plot(learned['step'], ema_smooth(learned[' val_loss']), label='Learned', color='orange')

plt.plot(none['step'], none[' val_loss'], alpha=0.15, color='green')
plt.plot(none['step'], ema_smooth(none[' val_loss']), label='None', color='green')

plt.plot(rotary['step'], rotary[' val_loss'], alpha=0.15, color='blue')
plt.plot(rotary['step'], ema_smooth(rotary[' val_loss']), label='Rotary', color='blue')

plt.plot(sinusoidal['step'], sinusoidal[' val_loss'], alpha=0.15, color='purple')
plt.plot(sinusoidal['step'], ema_smooth(sinusoidal[' val_loss']), label='Sinusoidal', color='purple')

plt.xlabel('Validation Step')
plt.ylabel('Validation Loss')
plt.title('Positional Encoding')
plt.legend()
plt.grid()
plt.savefig('positional_encoding.jpg')
plt.show()
```

```python
# Plotting all the validation accuracy curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(alibi['step'], alibi[' val_acc'], alpha=0.15, color='red')
plt.plot(alibi['step'], ema_smooth(alibi[' val_acc']), label='Alibi', color='red')

plt.plot(learned['step'], learned[' val_acc'], alpha=0.15, color='orange')
plt.plot(learned['step'], ema_smooth(learned[' val_acc']), label='Learned', color='orange')

plt.plot(none['step'], none[' val_acc'], alpha=0.15, color='green')
plt.plot(none['step'], ema_smooth(none[' val_acc']), label='None', color='green')

plt.plot(rotary['step'], rotary[' val_acc'], alpha=0.15, color='blue')
plt.plot(rotary['step'], ema_smooth(rotary[' val_acc']), label='Rotary', color='blue')

plt.plot(sinusoidal['step'], sinusoidal[' val_acc'], alpha=0.15, color='purple')
plt.plot(sinusoidal['step'], ema_smooth(sinusoidal[' val_acc']), label='Sinusoidal', color='purple')
plt.xlabel('Validation Step')
plt.ylabel('Validation Accuracy')
plt.title('Positional Encoding')
plt.legend()
plt.grid()
plt.show()
```
