```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in all the validation_results.log files from the different experiments
alibi = pd.read_csv('./alibi/validation_results.log')
learned = pd.read_csv('./learned/validation_results.log')
none = pd.read_csv('./none/validation_results.log')
rotary = pd.read_csv('./rotary/validation_results.log')
sinusoidal = pd.read_csv('./sinusoidal/validation_results.log')
```

```python
def ema_smooth(x, alpha=0.2):
    smoothed = []
    last = x[0]
    for v in x:
        last = alpha * v + (1 - alpha) * last
        smoothed.append(last)
    return smoothed
```

```python
# Plotting all the validation loss curves for comparison
# Use vaLidation step as x-axis and validation loss as y-axis
plt.figure(figsize=(12, 8))
plt.plot(alibi['step'], alibi[' val_loss'], label='Alibi', alpha=0.15, color='teal')
plt.plot(alibi['step'], ema_smooth(alibi[' val_loss']), label='Alibi (smoothed)', color='teal')
plt.plot(learned['step'], learned[' val_loss'], label='Learned', alpha=0.15, color='orange')
plt.plot(learned['step'], ema_smooth(learned[' val_loss']), label='Learned (smoothed)', color='orange')
plt.plot(none['step'], none[' val_loss'], label='None', alpha=0.15, color='purple')
plt.plot(none['step'], ema_smooth(none[' val_loss']), label='None (smoothed)', color='purple')
plt.plot(rotary['step'], rotary[' val_loss'], label='Rotary', alpha=0.15, color='green')
plt.plot(rotary['step'], ema_smooth(rotary[' val_loss']), label='Rotary (smoothed)', color='green')
plt.plot(sinusoidal['step'], sinusoidal[' val_loss'], label='Sinusoidal', alpha=0.15, color='red')
plt.plot(sinusoidal['step'], ema_smooth(sinusoidal[' val_loss']), label='Sinusoidal (smoothed)', color='red')
plt.xlabel('Validation Step')
plt.ylabel('Validation Loss')
plt.title('Validation Loss Comparison Across Different Optimizer Setups')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plotting all the validation accuracy curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(alibi['step'], alibi[' val_acc'], label='Alibi', alpha=0.15, color='teal')
plt.plot(alibi['step'], ema_smooth(alibi[' val_acc']), label='Alibi (smoothed)', color='teal')
plt.plot(learned['step'], learned[' val_acc'], label='Learned', alpha=0.15, color='orange')
plt.plot(learned['step'], ema_smooth(learned[' val_acc']), label='Learned (smoothed)', color='orange')
plt.plot(none['step'], none[' val_acc'], label='SGD', alpha=0.15, color='purple')
plt.plot(none['step'], ema_smooth(none[' val_acc']), label='None (smoothed)', color='purple')
plt.plot(rotary['step'], rotary[' val_acc'], label='Rotary', alpha=0.15, color='green')
plt.plot(rotary['step'], ema_smooth(rotary[' val_acc']), label='Rotary (smoothed)', color='green')
plt.plot(sinusoidal['step'], sinusoidal[' val_acc'], label='Sinusoidal', alpha=0.15, color='red')
plt.plot(sinusoidal['step'], ema_smooth(sinusoidal[' val_acc']), label='Sinusoidal (smoothed)', color='red')
plt.xlabel('Validation Step')
plt.ylabel('Validation Accuracy')
plt.title('Validation Accuracy Comparison Across Different Optimizer Setups')
plt.legend()
plt.grid()
plt.show()
```
