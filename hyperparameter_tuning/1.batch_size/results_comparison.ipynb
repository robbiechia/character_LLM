```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in all the validation_results.log files from the different experiments
batch16 = pd.read_csv('./batch_size_16/validation_results.log')
batch32 = pd.read_csv('./batch_size_32/validation_results.log')
batch64 = pd.read_csv('./batch_size_64/validation_results.log')
batch128 = pd.read_csv('./batch_size_128/validation_results.log')
```

```python
def ema_smooth(x, alpha=0.2):
    smoothed = []
    last = x[0]
    for v in x:
        last = alpha * v + (1 - alpha) * last
        smoothed.append(last)
    return smoothed
```

```python
# Plotting all the validation loss curves for comparison
# Use vaLidation step as x-axis and validation loss as y-axis
plt.figure(figsize=(12, 8))

plt.plot(batch16['step'], batch16[' val_loss'], label='batch_size_16', alpha=0.15, color='teal')
plt.plot(batch16['step'], ema_smooth(batch16[' val_loss']), label='batch_size: 16 (smoothed)', color='teal')

plt.plot(batch32['step'], batch32[' val_loss'], label='batch_size_32', alpha=0.15, color='orange')
plt.plot(batch32['step'], ema_smooth(batch32[' val_loss']), label='batch_size: 32 (smoothed)', color='orange')

plt.plot(batch64['step'], batch64[' val_loss'], label='batch_size_64', alpha=0.15, color='purple')
plt.plot(batch64['step'], ema_smooth(batch64[' val_loss']), label='batch_size: 64 (smoothed)', color='purple')

plt.plot(batch128['step'], batch128[' val_loss'], label='batch_size_128', alpha=0.15, color='green')
plt.plot(batch128['step'], ema_smooth(batch128[' val_loss']), label='batch_size: 128 (smoothed)', color='green')

plt.yscale('log')
plt.xlabel('Validation Step')
plt.ylabel('Validation Loss')
plt.title('Validation Loss Comparison Across Different Batch Size Setups')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plotting all the validation accuracy curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(batch16['step'], batch16[' val_acc'], label='batch_size: 16', alpha=0.15, color='teal')
plt.plot(batch16['step'], ema_smooth(batch16[' val_acc']), label='batch_size: 16 (smoothed)', color='teal')

plt.plot(batch32['step'], batch32[' val_acc'], label='batch_size: 32', alpha=0.15, color='orange')
plt.plot(batch32['step'], ema_smooth(batch32[' val_acc']), label='batch_size: 32 (smoothed)', color='orange')

plt.plot(batch64['step'], batch64[' val_acc'], label='batch_size: 64', alpha=0.15, color='purple')
plt.plot(batch64['step'], ema_smooth(batch64[' val_acc']), label='batch_size: 64 (smoothed)', color='purple')

plt.plot(batch128['step'], batch128[' val_acc'], label='batch_size: 128', alpha=0.15, color='green')
plt.plot(batch128['step'], ema_smooth(batch128[' val_acc']), label='batch_size: 128 (smoothed)', color='green')

plt.xlabel('Validation Step')
plt.ylabel('Validation Accuracy')
plt.title('Validation Accuracy Comparison Across Different Batch Size Setups')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plotting all the validation accuracy at the last character curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(batch16['step'], batch16[' last_char_val_acc'], label='batch_size: 16', alpha=0.15, color='teal')
plt.plot(batch16['step'], ema_smooth(batch16[' last_char_val_acc']), label='batch_size: 16 (smoothed)', color='teal')

plt.plot(batch32['step'], batch32[' last_char_val_acc'], label='batch_size: 32', alpha=0.15, color='orange')
plt.plot(batch32['step'], ema_smooth(batch32[' last_char_val_acc']), label='batch_size: 32 (smoothed)', color='orange')

plt.plot(batch128['step'], batch128[' last_char_val_acc'], label='batch_size: 128', alpha=0.15, color='purple')
plt.plot(batch128['step'], ema_smooth(batch128[' last_char_val_acc']), label='batch_size: 128 (smoothed)', color='purple')

plt.xlabel('Validation Step')
plt.ylabel('Validation Accuracy at Last Character')
plt.title('Validation Accuracy at Last Character Comparison Across Different Sequence Length Setups')
plt.legend()
plt.grid()
plt.show()
```
