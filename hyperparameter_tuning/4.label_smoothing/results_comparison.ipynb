```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in all the validation_results.log files from the different experiments
smoothing_0 = pd.read_csv('./smoothing_0/validation_results.log')
smoothing_0p1 = pd.read_csv('./smoothing_0p1/validation_results.log')
smoothing_0p05 = pd.read_csv('./smoothing_0p05/validation_results.log')
```

```python
def ema_smooth(x, alpha=0.2):
    smoothed = []
    last = x[0]
    for v in x:
        last = alpha * v + (1 - alpha) * last
        smoothed.append(last)
    return smoothed
```

```python
# Plotting all the validation loss curves for comparison
# Use vaLidation step as x-axis and validation loss as y-axis
plt.figure(figsize=(12, 8))

plt.plot(smoothing_0['step'], smoothing_0[' val_loss'], label='Label Smoothing: 0', alpha=0.15, color='teal')
plt.plot(smoothing_0['step'], ema_smooth(smoothing_0[' val_loss']), label='Label Smoothing: 0 (smoothed)', color='teal')

plt.plot(smoothing_0p1['step'], smoothing_0p1[' val_loss'], label='Label Smoothing: 0.1', alpha=0.15, color='orange')
plt.plot(smoothing_0p1['step'], ema_smooth(smoothing_0p1[' val_loss']), label='Label Smoothing: 0.1 (smoothed)', color='orange')

plt.plot(smoothing_0p05['step'], smoothing_0p05[' val_loss'], label='Label Smoothing: 0.05', alpha=0.15, color='purple')
plt.plot(smoothing_0p05['step'], ema_smooth(smoothing_0p05[' val_loss']), label='Label Smoothing: 0.05 (smoothed)', color='purple')

plt.yscale('log')
plt.xlabel('Validation Step')
plt.ylabel('Validation Loss')
plt.title('Validation Loss Comparison Across Label Smoothing Values')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plotting all the validation accuracy curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(smoothing_0['step'], smoothing_0[' val_acc'], label='Label Smoothing: 0', alpha=0.15, color='teal')
plt.plot(smoothing_0['step'], ema_smooth(smoothing_0[' val_acc']), label='Label Smoothing: 0 (smoothed)', color='teal')

plt.plot(smoothing_0p1['step'], smoothing_0p1[' val_acc'], label='Label Smoothing: 0.1', alpha=0.15, color='orange')
plt.plot(smoothing_0p1['step'], ema_smooth(smoothing_0p1[' val_acc']), label='Label Smoothing: 0.1 (smoothed)', color='orange')

plt.plot(smoothing_0p05['step'], smoothing_0p05[' val_acc'], label='Label Smoothing: 0.05', alpha=0.15, color='purple')
plt.plot(smoothing_0p05['step'], ema_smooth(smoothing_0p05[' val_acc']), label='Label Smoothing: 0.05 (smoothed)', color='purple')

plt.xlabel('Validation Step')
plt.ylabel('Validation Accuracy')
plt.title('Validation Accuracy Comparison Across Label Smoothing Values')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plotting all the validation accuracy at the last character curves for comparison
plt.figure(figsize=(12, 8))
plt.plot(smoothing_0['step'], smoothing_0[' last_char_val_acc'], label='Label Smoothing: 0', alpha=0.15, color='teal')
plt.plot(smoothing_0['step'], ema_smooth(smoothing_0[' last_char_val_acc']), label='Label Smoothing: 0 (smoothed)', color='teal')

plt.plot(smoothing_0p1['step'], smoothing_0p1[' last_char_val_acc'], label='Label Smoothing: 0.1', alpha=0.15, color='orange')
plt.plot(smoothing_0p1['step'], ema_smooth(smoothing_0p1[' last_char_val_acc']), label='Label Smoothing: 0.1 (smoothed)', color='orange')

plt.plot(smoothing_0p05['step'], smoothing_0p05[' last_char_val_acc'], label='Label Smoothing: 0.05', alpha=0.15, color='purple')
plt.plot(smoothing_0p05['step'], ema_smooth(smoothing_0p05[' last_char_val_acc']), label='Label Smoothing: 0.05 (smoothed)', color='purple')

plt.xlabel('Validation Step')
plt.ylabel('Validation Accuracy at Last Character')
plt.title('Validation Accuracy at Last Character Comparison Across Label Smoothing Values')
plt.legend()
plt.grid()
plt.show()
```
